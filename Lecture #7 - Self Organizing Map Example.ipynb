```python
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
```

```python
from sklearn.utils.extmath import cartesian

# SETUP 2D SPACE
x = np.linspace(0, 1, 40)
y = np.linspace(0, 1, 25)
nodes = cartesian([x,y])
np.random.shuffle(nodes)

# SETUP 3D SPACE
r = np.linspace(0, 1, 10)
g = np.linspace(0, 1, 10)
b = np.linspace(0, 1, 10)

colors = cartesian([r,g,b])
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(colors[:,0], colors[:,1], colors[:,2],c=colors)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(nodes[:,0],nodes[:,1],color=colors,s=300)
axes = plt.gca()
axes.set_xlim([np.min(nodes[:,0]),np.max(nodes[:,0])])
axes.set_ylim([np.min(nodes[:,1]),np.max(nodes[:,1])])    
fig.set_size_inches(18.5,8.5)
plt.show()
```

```python
for i in range(100000):
    random_color = np.random.uniform(0.0,1.0,(1,3))
    color_distances = cdist(colors,random_color,'euclidean')
    winning_node_index = np.argmin(color_distances)
    winning_node = np.array([nodes[winning_node_index]])
    node_distances = cdist(nodes,winning_node,'euclidean')
    gaussian = np.exp(-1*node_distances/0.03)
    update = (random_color-colors)*gaussian*0.3
    colors = colors + update
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(nodes[:,0],nodes[:,1],color=colors,s=300)
axes = plt.gca()
axes.set_xlim([np.min(nodes[:,0]),np.max(nodes[:,0])])
axes.set_ylim([np.min(nodes[:,1]),np.max(nodes[:,1])])    
fig.set_size_inches(18.5,8.5)
plt.show()
```
