```python
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
from sklearn.utils.extmath import cartesian
import csv
import matplotlib.patches as mpatches
import sys
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import plot, iplot, init_notebook_mode

fArr=[]
with open(r"C:\Users\manaskan\Desktop\PSU\fashion_mnist_train.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        if ('label' not in row[0]) and (int(row[0]) < 10):
             fArr.append(row)
```

```python
flist=list(fArr)
cArr=[]
cArray=[[230, 25, 75],[60, 180, 75],[255, 225, 25],[0, 130, 200],[245, 130, 48],[145, 30, 180],[70, 240, 240],[240, 50, 230],[250, 190, 190],[128, 0, 0]]
for idx,row in enumerate(flist):
    flist[idx]=list(flist[idx])+cArray[int(row[0])]
```

```python
fArray=np.array(flist)
np.random.shuffle(fArray)
nArray=fArray[0:1000]
dataArray=np.divide(fArray[:,1:785].astype(int),255)
nArray=np.asfarray(nArray,float)
srcSpace=fArray[:,1:785].astype(int)
srcSpace=np.divide(srcSpace,255)
```

```python
# SETUP datapoints
pixels=nArray[:,1:785]
pixelsBuf=np.divide(pixels,255)
colors=nArray[:,785:788]
colors=np.divide(colors,255)
```

```python
# SETUP 3D SPACE
r = np.linspace(0, 1, 10)
g = np.linspace(0, 1, 10)
b = np.linspace(0, 1, 10)

nodes = cartesian([r,g,b])
```

```python
color=np.multiply(colors,255)
c=[]
for col in color:
    c.append("rgb(%d,%d,%d)"%(col[0],col[1],col[2]))
trace2 = go.Scatter3d(
    x=nodes[:,0],
    y=nodes[:,1],
    z=nodes[:,2],
    mode='markers',
    marker=dict(
        color=c,
        size=10,
        symbol='circle'
    )
)
data = [trace2]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)
fig = go.Figure(data=data, layout=layout)
init_notebook_mode(connected=True)
iplot(fig, filename='simple-3d-scatter')
```

```python
def organize(nodes=nodes,pixels=pixelsBuf,p1=0.03,p2=0.3,pixeldistMetric='euclidean',nodedistMetric='euclidean'):
    print("Training Start")
    for i in range(100000):
        sys.stdout.write("\r" + str(i))
        sys.stdout.flush()
        random_idx=np.random.randint(0,60000)
        random_pixel=np.array(fArray[random_idx,1:785],ndmin=2).astype(int)
        random_pixel=np.divide(random_pixel,255)
        pixel_distances = cdist(pixels,random_pixel,pixeldistMetric)
        winning_node_index = np.argmin(pixel_distances)
        winning_node = np.array([nodes[winning_node_index]])
        node_distances = cdist(nodes,winning_node,nodedistMetric)
        gaussian = np.exp(-1*node_distances/p1)
        update = (pixels[winning_node_index]-pixels)*gaussian*p2
        pixels = pixels + update
    print("\r\nTraining Done!")
    
    print("Labeling Start")
    for idx,pixel in enumerate(pixels):
        sys.stdout.write("\r" + str(idx))
        sys.stdout.flush()
        pixel_distances = cdist(dataArray,np.array(pixel,ndmin=2),'euclidean')
        winning_datapoint = np.argmin(pixel_distances)
        colors[idx] = np.divide(fArray[winning_datapoint,785:788].astype(int),255)
        if(colors[idx]) not in np.divide(cArray,255):
            print("Incorrect Pixel value")
    print("\r\nLabeling Done!")

    #print("p1: %0.2f, p2: %0.2f, pixeldistMetric: %s, nodedistmetric: %s"%(p1,p2,pixeldistMetric,nodedistMetric))
    
    color=np.multiply(colors,255)
    c=[]
    for col in color:
        c.append("rgb(%d,%d,%d)"%(col[0],col[1],col[2]))

    trace2 = go.Scatter3d(
        x=nodes[:,0],
        y=nodes[:,1],
        z=nodes[:,2],
        mode='markers',
        marker=dict(
            color=c,
            size=10,
            symbol='circle'
        )
    )
    data = [trace2]
    layout = go.Layout(
        title="p1: %0.2f, p2: %0.2f, pixeldistMetric: %s, nodedistmetric: %s"%(p1,p2,pixeldistMetric,nodedistMetric),
        margin=dict(
            l=0,
            r=0,
            b=0,
            t=0
        )
    )
    fig = go.Figure(data=data, layout=layout)
    init_notebook_mode(connected=True)
    iplot(fig, filename='simple-3d-scatter')
```

Above you can see the results from my various experiments trying to modify numuerical parameters for the gaussian and update equations. Increasing the numerical value of the gaussian equation (which I think means including more of the neighbors into the picture), leads to converging to two or three labels. So I believe the immediate neighbors are the ones who carry a lot on weight.

Even with my best results I see multiple groupings for the same labels. This I think is a drawback for the SOM algorithm over T-SNE.

And since I used multiple cdist metrics like chebyshev, euclidean etc, I notice eucildean yields the best results. This tells me that in a 784 dimension spaces similar datapoints are closer to each other.
